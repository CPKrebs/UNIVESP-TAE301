# Cálculo Hash: Bitcoin

Bitcoin usa: SHA256(SHA256(Block_Header))

O Block_Header é composto por:

- Versão do bloco - 4 bytes
- Hash do bloco anterior - 32 bytes
- Hash de todas as transações no bloco - 32 bytes
- Carimbo do tempo - 4 bytes
- Bits (define dificuldade de mineração) - 4 bytes
- Nonce: a variavel que é incrementada - 4 bytes

O código hash só é aceito quando os 8 primeiros bytes forem iguais a 0



```python
import hashlib
from binascii import unhexlify, hexlify
header_hex = ("01000000" +
 "81cd02ab7e569e8bcd9317e2fe99f2de44d49ab2b8851ba4a308000000000000" +
 "e320b6c2fffc8d750423db8b1eb942ae710e951ed797f7affc8892b0f1fc122b" +
 "c7f5d74d" +
 "f2b9441a" +
 "42a14695")
header_bin = unhexlify(header_hex)
hash = hashlib.sha256(hashlib.sha256(header_bin).digest()).digest()
hexlify(hash[::-1]).decode("utf-8")
```
